```python
import numpy as np

np.random.seed(998877)

single_digit = np.random.randint(0, 10, size=(12))
eight_digit = np.random.randint(10000000, 99999999, size=(12))

print(single_digit)
print(eight_digit)



# G = [10, 8, 13, 8, 13, 10]
G = [11, 9, 12, 9, 12, 11]
N = len(G)
for i in range(N):
    l_or_g = i < N/2 # np.random.randint(2) # loss is 0, gain is 1
    g =  G[i] # np.random.randint(8, 15)
    endowment = None
    cert = None
    chance = None
    action = None

    if l_or_g == 1:
        endowment = 2
        cert = g
        chance = 2*g + 2
        action = "Recieve"
    else:
        endowment = 2*g + 4 
        cert = -g - 2
        chance = -2*g - 2
        action = "Lose"
        
    print(f"\nFor this decision, you have an Endowment of ${endowment}. Please choose one of the options below: ")
    print(f"\t Option A: {action} ${abs(cert)} for certain." )
    print(f"\t Option B: {action} ${abs(chance)} with 50% chance or {action} $0 with 50% chance." )
    
repeat = {}

np.random.seed(92299)
F = [105, 110, 115, 125]

T = ["Today", "In One Week", "In One Month"]

t = ["Day", "Week", "Month"] 

i = 0
while i < 6:
    a = np.random.randint(0, 3)
    b = None
    if a == 2:
        b = np.random.randint(0, 2)
    else:
        b = np.random.randint(0, 3)
    c = np.random.randint(0, 3)

    temp = str(a) + str(b) + str(c)
    if temp in repeat.keys():
        continue
    else:
        repeat[temp] = ''
        i += 1

    print("\nWhich of the following do you prefer to have?")
    print(f"\t Option A: $100 {T[a]}.")
    if a == 0 and a == b:
        print(f"\t Option B: ${F[c]} Tomorrow.")
    else:
        print(f"\t Option B: ${F[c]} {T[a]} plus a {t[b]}.")
```
